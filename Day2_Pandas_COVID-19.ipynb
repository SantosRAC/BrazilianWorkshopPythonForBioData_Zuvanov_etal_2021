1.1 Introduction to the Pandas Module

`pandas` is a Python Module widely used for it's data structures, quick and flexible tools for data analysis. These structures' main goal is to make relational data manipulation easy and intuitive.

To use the pandas module, it needs to be imported via the command `import`. We can also rename the modules to an easier name; that way, we won't have to type it's full name whenever we want to use. It's common to import pandas like the following:

```
# Discuss what importing a module means
import numpy as np
import pandas as pd
import scipy.stats as st
import matplotlib.pyplot as plt
```

```
help(pd)
```

Thw two main data structures in `pandas` are `Series` and `DataFrame`. The first is a general data structure to work with single-dimension data, while the second one is used to work in two dimensions.

`Series` are basically lists of values that are capable of storing any type of data inside of them, like integers, *strings*, *floats*, etc. The position of every object in a `Series` is called the index. By default, the indexes are integers in ascending order `starting at value 0` (just like in `sequences`).

We can build a `Series` object using the `.Series` command from the module.

A `DataFrame` is a two-dimensional data structure, in other words, the data is organized in rows and columns. In this case, every column can hve a different data type. A DataFrame can be created from a list or multiple lists of values and, just like the Series, the DataFrames have indexes that represent rows and columns.

A big advantage of the DataFrames is that they have most of the same functions that the great `data.frames` from the R language, making it easier for users of that languange to work with pandas.

```
# Apply multiple functions for each group
dfGPsumm = dfGP.groupby(by = ['Treatment', 'Crop'])['Value'].agg(['count', 'mean', 'sem'])
dfGPsumm
```

```
# Calculate the confidence interval (using the t distribution, that would be the more correct one)
# Product element by element
alpha = 0.05
dfGPsumm['confint'] = dfGPsumm['sem'] * st.t.ppf(1 - alpha/2, df = dfGPsumm['count'] - 1)
dfGPsumm
```

```
# To plot a figure, the DataFrame needs to be in the wide format
# (each column represents a Series in the figure)
dfGPsumm = dfGPsumm.unstack(1)
dfGPsumm
```
