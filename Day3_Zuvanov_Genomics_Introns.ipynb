<br>

Are genetics and genomics synonymous? Although both names refer to the study of genes, the areas of activity and interests can be quite different for these two areas of knowledge. In short, genetics "is the study of heredity" and genomics "is the study of genes and their functions, and related techniques" (__[WHO definitions of Genetics and Genomics](https://www.who.int/genomics/geneticsVSgenomics/en/)__). Thus, while genetics "examines the function and composition of an individual gene, genomics studies all the genes in an organism and their interrelationships in order to identify their influence together on the growth and development of that organism" (__[WHO definitions of Genetics and Genomics](https://www.who.int/genomics/geneticsVSgenomics/en/)__). To better understand these disciplines, let us recall how and when these studies emerged. For that, it’s worth taking a look at Phil McClean’s work entitled "__[A History of Genetics and Genomics](https://www.ndsu.edu/pubweb/~mcclean/plsc411/History-of-Genetics-and-Genomics-narrative-and-overheads.pdf)__" where we got all the historical context information.  

Our history begins in the middle of the 19th century, a period in which the search for an explanation of the origin and variability of effervescent species. In 1858, Darwin and Wallace made history by describing how "species arise via evolution and how natural selection uses natural variation to develop new forms". Later, Mendel publishes a work that describes the "hereditary unit as a particle that does not change". This idea differed from the hitherto model of inheritance by mixing paternal characteristics. While Mendel was working on the idea of ​​quantized inheritance, several other discoveries were made in isolation: "Haeckel predicted that hereditary material is found in the nucleus"; "Miescher showed that the nucleus material was nucleic acids" and "others observed the behavior of chromosomes and suggested that they had a role in heredity". As McClen said, "imagine how the concepts could have evolved if the information were as widespread as it is today"! 

In the early 20th century, Mendel's work was <i> rediscovered </i> and with it Mendel's famous Laws. Do you remember what they are? After this rediscovery, some genetic principles were described, "as <i> genetic linkage </i>, lethal genes and maternal inheritance". It was also at the beginning of the century that there were great studies and discoveries on the physical basis of heredity that were synthesized in the "chromosomal theory of inheritance". In this theory, Sutton and Boveri suggested that it is on the chromosome that the hereditary unit is located (cytogenetics is based on this theory, do you know what it studies?). Chromosomal anomalies and the effect of gene positioning on the chromosome have shown that there is a "direct link between chromosomes and the phenotype". 

However, it was only in the mid-twentieth century, through various experiments (from Griffith, Avery, MacLeod and McCarty, and from Hershey and Chase, do you know any?), That we really look at DNA as genetic material. Time later - using X-ray diffractions made by Rosalind Franklin, Maurice Wilkins and Raymond Gosling and also the discovery of DNA pairing in 1: 1 ratio by Erwin Chargaff - scientists Watson and Crick determined the structure of DNA. Soon after, the "central dogma of molecular biology" was developed (what is this <i> dogma </i>, is it correct?). It was also in this century that Mendel and Darwin's theories were merged to form what became known as "Neo-Darwininsmo". In this "modern synthesis", it is said that "mutations create variation; recombination develops new forms; variation extends across the population; and based on environmental restrictions, forces of natural selection finally act on variation to produce more adapted individuals".
At the end of the twentieth century, studies focused on the development of "molecular biology techniques, recombinant DNA technology, and the biotechnology industry" (which techniques can you remember?). "Advances in protein sequencing and DNA sequences have launched a new era of phylogenetics", allowing species to be compared at the molecular level. At this point, there is also the "neutralist theory of evolution" that suggests that most mutations prevail for reasons other than natural selection, such as genetic drift. 

The sequencing of the first complete genome (bacteriophage phiX174 from <i> E. coli </i> by the Sanger group) and then several other organisms (such as the Human Genome Project), were fuel for the development of bioinformatics! The famous "omica revolution", amid the great increase in biological data and the information age (with the emergence of the WWW and e-mail), made it possible to study biological molecules, their functions and dynamics in order to seek understanding phenomena in their entirety.

#### - GFF File

With only the FASTA file, would you be able to tell where the genes, exons, introns, *splicing* sites or any other characteristic of the genome are? Difficult right? The goal now is to show how we can find regions of interest in the chromosomes of a genome. We need to have an additional document, which is the genome annotation file.

There are two main formats, *General Feature Format* (GFF) and *General Transfer Format* (GTF). In GFF, each annotation will be arranged on a line and there are columns separated by tabs, which represent:

 * Sequence name (seqname): Name of the chromosome or *scaffold* where this annotation is located;
 * Source: Name of the program or source that generated the information;
 * Biological type (type): biological type of this annotation
   * gene;
   * transcript;
   * exon;
   * CDS;
   * five_prime_utr;
   * three_prime_utr;
   * start_codon;
   * stop_codon.

 * Beginning of the location of this annotation in the genome (start);

 * End of the location of this annotation in the genome (end);

 * Score;

 * Strand: Indicates whether the annotation is on the positive (+) or negative (-) strand;

 * Reading frame (phase): Indication of the start position of the annotation codon. It can take the values '0', '1', '2' or '.';

 * Attribute: Additional information about the annotation.

Attention: Empty columns are marked by a '.', Do not be alarmed by the number of them!

- create a **figure** using the *figure()* function;
- add a **subplot** using the function *add_subplot()*;
- plot an **Axes** using one of the different plot functions, such as *boxplot()*, *scatter()*, *lines()* ...
